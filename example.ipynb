## Building data and precalculated energy demand values from the EnergyMap Berlin data base

Buildings within a polygon

```python
from energymap4py import data_access as em_data
test_poly = [(13.322027104721627, 52.51773514881992),
             (13.321124828370177, 52.516709837521866),
             (13.323525466163757, 52.51597401976454),
             (13.324266491874166, 52.5169650913811),
             (13.322847600680866, 52.517782208350184)]

res = em_data.by_polygon(test_poly)
for i in res:
    print(i)
```

One building defined by a building UUID

```python
res = em_data.by_uuid('DEBE04YY500002vO')
for i in res:
    print(i)
```

A couple of buildings 25 m right and left of a polyline

```python
test_line = [(13.324412829321979, 52.516957033745285),
             (13.322429034685205, 52.514576803927426),
             (13.322485599381766, 52.51321707446597)]

res = em_data.by_line(line_points=test_line, dist=25)
for i in res:
    print(i)
```

A couple of buildings within a circle defined by a point (longitiude, latitide) and a radis of 50 m

```python
test_lon = 13.467488494175514
test_lat = 52.50587905804428

res = em_data.by_point(lon=test_lon, lat=test_lat, dist=50)
for i in res:
    print(i)
```
